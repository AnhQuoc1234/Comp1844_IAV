```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
# Create the graph
MyGraph = nx.Graph()

# Stations and distances
edges_with_distances = [
    ("Hyde Park Corner", "Green Park", 0.3),
    ("Green Park", "Piccadilly Circus", 0.4),
    ("Piccadilly Circus", "Leicester Square", 0.2),
    ("Leicester Square", "Covent Garden", 0.1),
    ("Covent Garden", "Holborn", 0.3)
]

# Add nodes, edges and weights
for station1, station2, distance in edges_with_distances:
    MyGraph.add_edge(station1, station2, weight=distance)

# Positions for each station
pos = {
    "Hyde Park Corner": (0, 0),
    "Green Park": (1, 1),
    "Piccadilly Circus": (2, 1),
    "Leicester Square": (3, 1),
    "Covent Garden": (4, 1.5),
    "Holborn": (5, 2)
}

# Offset for label positioning
label_offset = 0.2

# Draw the network
plt.figure(figsize=(12, 10))
nx.draw(MyGraph, pos, with_labels=False, node_size=500, node_color="blue", font_size=10, font_color="black", edge_color="black", width=1)

# For station names closer to the nodes
for station, (x, y) in pos.items():
    plt.text(x, y - 0.1, station, fontsize=10, ha='center', va='center', color="black")

# Add edge labels
edge_labels = {(station1, station2): f"{distance} miles" for station1, station2, distance in edges_with_distances}
nx.draw_networkx_edge_labels(MyGraph, pos, edge_labels=edge_labels, font_color="black")

# Add legend 
plt.plot([], [], color="black", linewidth=1, label="Piccadilly Line") 
plt.legend(loc="lower right", fontsize=10)

plt.title("Piccadilly Line")
plt.show()
```
